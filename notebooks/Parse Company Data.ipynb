# Purpose

This notebook parses the company data stored in `../Companies.xlsx` and converts it into a JSON dictionary that is used by the final script.

```python
import pandas as pd
import json as json
```

```python
companies_df = pd.read_excel('../Companies.xlsx')

companies_df
```

```python
companies_search_config = dict()

for _, s in companies_df.iterrows():
    company_search_config = list()
    
    # Add "Name" to search list
    company_search_config.append(s['Name'].lower())
    
    # Add aliases to search list
    for alias in s['Aliases'].split(','):
        company_search_config.append(alias.strip())
    
    # Add to main config dictionary
    companies_search_config[s['Symbol']] = tuple(company_search_config)
    
companies_search_config
```

```python
with open('../dse203_final_project/companies.json', 'w') as f:
    json.dump(companies_search_config, f, indent=4)
```
